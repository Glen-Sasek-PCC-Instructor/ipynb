# Scatter Plot

Scatter Plots will show relationships between two columns.

Choose 2 numeric columns, or one text and one numeric to display a relationship between them.

Remember to add strings for the title, you can leave the axis labels blank.

```python
# Plot the graph and add labels
row="tempo"
col="release_year"
df.plot.scatter(x=col, y=row, figsize=(14,8))

# Add a title and axis labels
plt.xlabel("")
plt.ylabel("")
plt.title("")

# Uncomment the following statements to draw horizontal or vertical lines.
# Use axvline() for vertical lines.
# plt.axhline(y=-.22, color="r")
# plt.axhline(y=.78, color="r")
```

# Crosstab (Cross-Tabulation) Chart

Crosstab Charts are used to display counts of distict values in one column mapped to values in another column. The chart is displayed in a table format. Only use when the number of distinct values is small, otherwise you will generate a huge table!

You can apply a heatmap to your table to show the higher values. Try changing the color name to find one that works best.

Choose 2 columns to create a crosstab chart. Update the the column name for rows in the 'index' argument, and the other column in the 'columns' argument.

Add a string to the first print statement to create a title.

```python
# Add a title to the chart using Python print() statement
print('\n')

# Adding a heatmap to highlight the higher numbers, you must import
# the import seaborn libarary: import seaborn as sns
orange = sns.light_palette("orange", as_cmap=True)

# Create the crosstab chart, specify two column names: index = the rows, and
# columns = the columns.
pd.crosstab(index=df[row], columns=[df[col]]).style.background_gradient(cmap=orange)
```

```python
# Use this code to create a crosstab chart without a heatmap to display total counts.

pd.crosstab(index=df[row], columns=[df[col]], margins=True)
```
