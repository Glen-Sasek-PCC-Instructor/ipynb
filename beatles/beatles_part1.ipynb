# Bar Chart
Bar charts will display the counts of distinct values in a column.

Choose a column you want and update the column_name variable in the first statement. The graph will display the counts for each of the distinct values in the column, ordered by the x-axis. If you want the graph sorted by counts from highest to lowest, comment out the second statement:

`#bar_df = bar_df.sort_values('index')`

Remember to add strings for the title and x axis label.

```python
# set this variable to the column name you want to graph
# (e.g. 'genre')
column_name = 'genre'

bar_df = df[column_name].value_counts().to_frame('count').reset_index()
bar_df = bar_df.sort_values(column_name)
bar_df.plot.bar(x=column_name, figsize=(14, 8))

# Add a title and axis labels
plt.xlabel("")
plt.ylabel("Count")
plt.title("")
```

# Donut Chart

Donut charts will display the percentages of counts. Donut charts are really Pie charts with a white circle drawn on top.

Choose the column you used for the Bar Chart and change the column_nmae variable int the first statement. You can comment out the statements to draw the white circle if you want to display a Pie Chart.

Remember to add strings for the title, you can leave the axis labels blank.

```python
# set this variable to the column name you want to graph
# (e.g. 'genre')
column_name = 'genre'

donut_df = df[column_name].value_counts()
donut_df.plot.pie(figsize=(14, 7), autopct='%1.1f%%', subplots=True)

# The next 3 statements add a white circle to draw a donut chart
my_circle = plt.Circle( (0,0), 0.5, color='white')
p = plt.gcf()
p.gca().add_artist(my_circle)

# Add a descriptive title
plt.title(" ")
```

# Histogram

Histograms are used with one column of numeric data when there are too many distinct values to graph using a bar chart. Histrgrams create 'bins' to display counts of values that are within the range of numbers the bin represents.

Choose a numeric column with many distinct values. Adjust the bin size until you create an interesting graph.

Remember to add strings for the title and axis labels.

```python
column_name = 'release_year'
df[column_name].plot.hist(figsize=(14,8), rwidth=0.9, bins=10)

# Add the title and axis labels
plt.title(" ")
plt.xlabel(" ")
plt.ylabel("Counts")
```
