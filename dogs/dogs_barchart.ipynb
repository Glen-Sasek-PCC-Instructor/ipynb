```python
# Create a dataframe from our csv file
df = pd.read_csv("dogs.csv")
```

```python
df.columns
```

# Bar Chart

```python

index='Max Life Span'
# Create a new dataframe called bar_df with each ‘Max Life Span’ value and the number of times (counts) that value appears in the column
bar_df = df[index].value_counts().to_frame('count').reset_index()

# Sort the dataframe by ‘index’, which is the ‘Max Life Span’ column
bar_df = bar_df.sort_values(index)

# Plot the dataframe and use the column ‘index’ for x-axis
bar_df.plot.bar(x=index, figsize=(14, 8))

#Add labels to the chart
plt.xlabel("# Years")
plt.ylabel("Count")
plt.title("Count of Max Life Spans")
```
