# Cross-tabulation (Crosstab) Chart

```python
# Add a title to the chart using Python print() statement
row='Max Weight'
column='Breed Group'

print('\t',column, ' per ', row, '\n')

# Create the crosstab chart, specify two column names: index = the rows, and
# columns = the columns.
pd.crosstab(index=df[row], columns=[df[column]], margins=True)
```

# Crosstab Chart with Heatmap

Importing the Seaborn library, you can use functions to draw a heatmap, which
darkens the cells with higher values.

```python

# Add a title to the chart using Python print() statement
print('Maximum Life Span Counts per Breeding Group\n')

# Adding a heatmap to highlight the higher numbers, you must import
# the import seaborn libarary: import seaborn as sns
orange = sns.light_palette("orange", as_cmap=True)

# Create the crosstab chart, specify two column names: index = the rows, and
# columns = the columns.
pd.crosstab(index=df['Breed Group'], columns=[df['Max Life Span']]).style.background_gradient(cmap=orange)
```
