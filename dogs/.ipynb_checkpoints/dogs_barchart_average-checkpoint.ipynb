# **Dogs Dataset**

```python
import matplotlib.pyplot as plt
import pandas as pd
```

Download the [dogs.csv](https://drive.google.com/file/d/1JB7i-u7q3YpuxihimQG5GiGTKac-_MLj/view?usp=sharing) file onto your computer and drag into this Notebook.

```python
# Create a dataframe from our csv file
df = pd.read_csv("dogs.csv")
```

```python
df.columns
```

# Bar Chart to display Sum or Mean (Average)

```python
# Create a new dataframe, group all of the dogs together by
# Breed Group and calculate the mean of all the numeric columns
mean_df = df.groupby(['Breed Group'])['Max Weight'].mean() # to get the sum
                                             # use .sum()

# Plot the bar chart
mean_df.plot.bar(figsize=(14, 8))
plt.xlabel("Breed Group")
plt.ylabel("Average Max Weight (lbs)")
plt.title("Average Max Weight per Breed Group")
```

# Donut Chart (Pie Chart)

You can play around with the code below to plot another useful chart to display counts as percentages of the whole - Pie Chart or Donut Chart:

```python
# Create a Pie Chart
donut_df = df['Max Life Span'].value_counts()
donut_df.plot.pie(figsize=(14, 7), autopct='%1.1f%%', subplots=True)

# The next 3 statements add a white circle to draw a donut chart
my_circle=plt.Circle( (0,0), 0.5, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)

# Add the labels and title
plt.xlabel("# Years")
plt.ylabel(" ")
plt.title("% of Max Life Spans")
```
