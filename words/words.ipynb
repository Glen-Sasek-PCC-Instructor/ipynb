```python
# Crosstab without Heatmap

# Add a title to the chart using Python print() statement
print('Word Length Counts per Part of Speech\n')

pd.crosstab(index=df['Length'], columns=[df['Part of Speech']], margins=True)
```

```python
# Crosstab with Heatmap

# Add a title to the chart using Python print() statement
print('Word Length Counts per Part of Speech\n')

# Adding a heatmap to highlight the higher numbers, you must import
# the import seaborn libarary: import seaborn as sns
green = sns.light_palette("green", as_cmap=True)

# Create the crosstab chart, specify two column names: index = the rows, and
# columns = the columns.
pd.crosstab(index=df['Length'], columns=[df['Part of Speech']]).style.background_gradient(cmap=green)
```
